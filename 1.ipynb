# **Grammar & Language in NLP**

```python
import re

from nltk import TreebankWordTokenizer

# simple regular grammar using a regex pattern

pattern = r"\b\w+\b" # Matches a word
text = "We are going to build an AI model"
tokens = re.findall(pattern, text)
print(tokens)
```

*   \b → Matches a word boundary, ensuring that the word is separate from surrounding non-word characters (like spaces, punctuation, etc.).
*   \w+ → Matches one or more word characters ([A-Za-z0-9_]).
*   \b → Ensures that the match ends at a word boundary.

```python
def fsa_token_recognizer(word):
    state= 0
    for char in word:
        if state == 0 and char.isalpha():
            state = 1
            print(f"State: {state} - Character: {char}")
        elif state == 1 and char.isalpha():
            state = 1
            print(f"State: {state} - Character: {char}")
        else:
            print(f"State: {state} - Character: {char}")
    return state == 1
```

## custom-defined Context-Free Grammar (CFG) with NLTK's parser

### **NLTK is a leading platform for building Python programs to work with human language data.**

#### It provides easy-to-use interfaces to over 50 corpora and lexical resources


```python
import nltk

# Revised grammar with adjective support and defined PP
grammar = nltk.CFG.fromstring("""
S -> NP VP
NP -> Det N
NP -> Det Adj N
NP -> Det N PP
NP -> Det Adj N PP
VP -> V NP
VP -> V
PP -> P NP
Det -> 'the' | 'a'
Adj -> 'big' | 'small'
N -> 'cat' | 'dog'
V -> 'chases' | 'sleeps'
P -> 'with' | 'in'
""")

parser = nltk.ChartParser(grammar)
sentence = ['the', 'big', 'cat', 'chases', 'a', 'small', 'dog']

for tree in parser.parse(sentence):
    tree.pretty_print()
```
