```python
import spacy
import language_tool_python
!python -m spacy download en_core_web_sm
# set seeds for reproducibility
import random
import numpy as np
import torch
import os
os.environ['PYTHONHASHSEED'] = str(0)
random.seed(0)
np.random.seed(0)
torch.manual_seed(0)
torch.cuda.manual_seed(0)
torch.backends.cudnn.deterministic = True
torch.backends.cudnn.benchmark = False
```

```python

# Φόρτωση του Αγγλικού μοντέλου
nlp = spacy.load("en_core_web_sm")
tool = language_tool_python.LanguageTool('en-US')

def custom_reconstruct(sentence):
    print("\n--- Original Sentence ---")
    print(sentence)

    # Ανάλυση με spaCy
    doc = nlp(sentence)

    # Εκτύπωση lemmatization και POS tagging
    print("\nTokens | POS | Lemma:")
    for token in doc:
        print(f"{token.text:15} | {token.pos_:10} | {token.lemma_}")

    # Διόρθωση γραμματικής με language_tool
    grammar_corrected = tool.correct(sentence)
    print("\nGrammar Corrected:", grammar_corrected)

    # Κανόνες 
    final = grammar_corrected.replace(
        "as his next contract checking",
        "as he reviews his next contract"
    )

    return final

# Δείγματα από το Κείμενα 1 και 2
sentence_1 = "Today is our dragon boat festival, in our Chinese culture, to celebrate it with all safe and great in our lives."
sentence_2 = "Anyway, I believe the team, although bit delay and less communication at recent days, they really tried best for paper and cooperation."

# Εφαρμογή
reconstructed_1 = custom_reconstruct(sentence_1)
reconstructed_2 = custom_reconstruct(sentence_2)
# Αποτελέσματα
print("\n=== Final Reconstructed Sentences ===")
print("1:", reconstructed_1)
print("2:", reconstructed_2)
```

```python
from transformers import pipeline
import language_tool_python
from textblob import TextBlob

# BART paraphrasing
paraphraser = pipeline("text2text-generation", model="facebook/bart-large-cnn")

def apply_bart(text):
    result = paraphraser(text, max_length=512, do_sample=False)[0]['generated_text']
    return result

# LanguageTool grammar correction
tool = language_tool_python.LanguageTool('en-US')

def apply_language_tool(text):
    return tool.correct(text)

# TextBlob simplification and correction
def apply_textblob(text):
    blob = TextBlob(text)
    return str(blob.correct())

# Εφαρμογή και στα δύο κείμενα
def process_all(text1, text2):
    print("=== Pipeline 1: BART ===")
    print("\nText 1:\n", apply_bart(text1))
    print("\nText 2:\n", apply_bart(text2))

    print("\n=== Pipeline 2: LanguageTool ===")
    print("\nText 1:\n", apply_language_tool(text1))
    print("\nText 2:\n", apply_language_tool(text2))

    print("\n=== Pipeline 3: TextBlob ===")
    print("\nText 1:\n", apply_textblob(text1))
    print("\nText 2:\n", apply_textblob(text2))

# Εκτέλεση
text1 = "Today is our dragon boat festival, in our Chinese culture, to celebrate it with all safe and great in our lives. Hope you too, to enjoy it as my deepest wishes. Thank your message to show our words to the doctor, as his next contract checking, to all of us. I got this message to see the approved message. In fact, I have received the message from the professor, to show me, this, a couple of days ago. I am very appreciated the full support of the professor, for our Springer proceedings publication"

text2 = "During our final discuss, I told him about the new submission — the one we were waiting since last autumn, but the updates was confusing as it not included the full feedback from reviewer or maybe editor? Anyway, I believe the team, although bit delay and less communication at recent days, they really tried best for paper and cooperation. We should be grateful, I mean all of us, for the acceptance and efforts until the Springer link came finally last week, I think. Also, kindly remind me please, if the doctor still plan for the acknowledgments section edit before he sending again. Because I didn’t see that part final yet, or maybe I missed, I apologize if so. Overall, let us make sure all are safe and celebrate the outcome with strong coffee and future targets"

process_all(text1, text2)
```

## Σύγκριση Αλγορίθμων

| Αλγόριθμος    | Είδος Αλλαγής           | Παράδειγμα Αλλαγής                 | Πλεονεκτήματα                              | Μειονεκτήματα                             |
|---------------|-------------------------|------------------------------------|--------------------------------------------|-------------------------------------------|
| BART          | Παραφράσεις / Περίληψη  | Ολόκληρες προτάσεις αφαιρέθηκαν    | Φυσική ροή, βελτιωμένη σύνταξη             | Αφαίρεση πληροφορίας                      |
| LanguageTool  | Γραμματική διόρθωση     | was → were                         | Ακριβείς γραμματικές αλλαγές               | Δεν αλλάζει ύφος, δεν διορθώνει ύφος      |
| TextBlob      | Ορθογραφικός έλεγχος    | Springer → Printer                 | Αυτόματες διορθώσεις σε απλές λέξεις       | Χαμηλή ακρίβεια, αλλαγή κρίσιμων λέξεων   |

## Συμπέρασμα

Ο αλγόριθμος BART είναι αποτελεσματικός σε επίπεδο ύφους και ροής, αλλά ενδέχεται να παραλείψει πληροφορίες.  
Το LanguageTool είναι ασφαλές για συντηρητικές διορθώσεις χωρίς παραφράσεις.  
Το TextBlob δεν είναι κατάλληλο για τεχνικά ή επιστημονικά κείμενα, λόγω ανακριβών διορθώσεων.
